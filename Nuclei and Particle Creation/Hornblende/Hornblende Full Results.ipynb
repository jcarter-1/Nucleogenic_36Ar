```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt

"""
Plotting
"""
import matplotlib as mpl
mpl.rcParams['font.size'] = 14
```

```python
"""
Hornblende Composition
"""
comps = np.array([4.66, 10.86, 17.26, 
                 49.62, 11, 2.4, 2.19, 
                 0.01, 2])

comps_label = ['K$_{2}$O', 'Al$_{2}$O$_{3}$', 
               'MgO', 
              'SiO$_{2}$',
               'CaO',
               'Fe$_{2}$O$_{3}$',
               'H$_{2}$O', 
               'Ar', 
              'Cl']
```

```python
fig, ax = plt.subplots(1,1, figsize = (4.2, 3.78))

ax.bar(np.arange(comps.shape[0])*1.1, comps, 
           edgecolor = 'k', alpha = 0.67, 
       color = 'RoyalBlue');

ax.bar(np.arange(comps.shape[0])*1.1, comps, 
           edgecolor = 'k', alpha = 1, lw = 1.1,
       facecolor = 'None'); 

ax.set_xticks(np.linspace(0, comps.shape[0] - 1,comps.shape[0])*1.1)
ax.set_xticklabels(comps_label, rotation = 45, 
                   fontsize = 12)
ax.set_ylabel('log(Fractional Mass ($\%$))')
ax.set_yscale('log')
ax.set_xlim(-0.7, (comps.shape[0]*1.1) - 0.4)
plt.tight_layout()
```
